## **Part 5 â€“ Net Pay Calculation**
 Objective: Combine deductions into final result.

1. Compute:
```
total_deductions = income_tax + health_levy + district_tax + education_deduction
net_pay = gross_pay âˆ’ total_deductions
```
2. Print:
- Gross Pay
- All deductions (with labels)

````python
# === Payroll Calculator ===

# Hourly rate
rate = 5

# User inputs
hours = float(input("Enter hours worked: "))
children = int(input("Enter number of children: "))

# --- Gross Pay Calculation ---
if hours <= 40:
    gross_pay = hours * rate
else:
    regular_pay = 40 * rate
    overtime_pay = (hours - 40) * (rate * 1.5)
    gross_pay = regular_pay + overtime_pay

# --- Deductions ---
income_tax = 0.15 * gross_pay
health_levy = 0.025 * gross_pay
district_tax = 0.01 * gross_pay

# Educational fund deduction
if children <= 3:
    education_deduction = 0
else:
    education_deduction = (children - 3) * 0.5

# --- Totals ---
total_deductions = income_tax + health_levy + district_tax + education_deduction
net_pay = gross_pay - total_deductions

# --- Output ---
print("\n=== PAYSLIP SUMMARY ===")
print(f"Hours Worked:          {hours}")
print(f"Number of Children:    {children}")
print(f"Gross Pay:             â‚¦{gross_pay:.2f}\n")

print("DEDUCTIONS:")
print(f"  Income Tax (15%):          â‚¦{income_tax:.2f}")
print(f"  Health Levy (2.5%):        â‚¦{health_levy:.2f}")
print(f"  District Tax (1%):         â‚¦{district_tax:.2f}")
print(f"  Education Deduction:       â‚¦{education_deduction:.2f}")
print(f"------------------------------------")
print(f"Total Deductions:            â‚¦{total_deductions:.2f}")
print(f"Net Pay:                     â‚¦{net_pay:.2f}")
````

## **Part 6 â€“ Multiple Employees (While Loop)**
 Objective: Practice repetition using a `while` loop.

1. Allow the program to handle **many employees**:
- Keep asking for employee details (hours, children).
- Stop only when the user enters a sentinel value (e.g., "done" for name).
2. For each employee, display their payroll breakdown.

ðŸ”¹ **Mini Task**: Enter payroll data for 3 employees, then type "done".

````python
# === MULTIPLE EMPLOYEE PAYROLL CALCULATOR ===

rate = 5  # hourly rate

def calculate_gross_pay(hours):
    if hours <= 40:
        return hours * rate
    else:
        regular_pay = 40 * rate
        overtime_pay = (hours - 40) * (rate * 1.5)
        return regular_pay + overtime_pay

def calculate_deductions(gross_pay, children):
    income_tax = 0.15 * gross_pay
    health_levy = 0.025 * gross_pay
    district_tax = 0.01 * gross_pay
    education_deduction = 0 if children <= 3 else (children - 3) * 0.5
    total_deductions = income_tax + health_levy + district_tax + education_deduction
    net_pay = gross_pay - total_deductions
    return income_tax, health_levy, district_tax, education_deduction, total_deductions, net_pay


print("=== PAYROLL SYSTEM ===")
print("Enter 'done' to stop.\n")

while True:
    name = input("Enter employee name (or 'done' to finish): ").strip()
    if name.lower() == "done":
        break

    hours = float(input(f"Enter hours worked for {name}: "))
    children = int(input(f"Enter number of children for {name}: "))

    # Calculations
    gross_pay = calculate_gross_pay(hours)
    income_tax, health_levy, district_tax, education_deduction, total_deductions, net_pay = calculate_deductions(gross_pay, children)

    # Display payslip
    print(f"\n=== PAYSLIP for {name} ===")
    print(f"Hours Worked:          {hours}")
    print(f"Number of Children:    {children}")
    print(f"Gross Pay:             â‚¦{gross_pay:.2f}\n")

    print("DEDUCTIONS:")
    print(f"  Income Tax (15%):          â‚¦{income_tax:.2f}")
    print(f"  Health Levy (2.5%):        â‚¦{health_levy:.2f}")
    print(f"  District Tax (1%):         â‚¦{district_tax:.2f}")
    print(f"  Education Deduction:       â‚¦{education_deduction:.2f}")
    print(f"------------------------------------")
    print(f"Total Deductions:            â‚¦{total_deductions:.2f}")
    print(f"Net Pay:                     â‚¦{net_pay:.2f}")
    print("====================================\n")

print("âœ… Payroll processing completed!")
````

## **Final Challenge â€“ Full Payroll Program**
ðŸŽ¯ Write a complete Python program that:
- Uses input to collect employee name, hours worked, and number of children.
- Calculates gross pay, all deductions, and net pay.
- Displays a formatted payroll summary for each employee.
- Allows multiple employees until "done" is typed.

````python
# === FULL PAYROLL PROGRAM ===
# Author: (Your Name)
# Objective: Calculate gross pay, deductions, and net pay for multiple employees.

rate = 5  # hourly rate in Naira

def calculate_gross_pay(hours):
    """Calculate gross pay including overtime"""
    if hours <= 40:
        return hours * rate
    else:
        regular_pay = 40 * rate
        overtime_pay = (hours - 40) * (rate * 1.5)
        return regular_pay + overtime_pay


def calculate_deductions(gross_pay, children):
    """Compute all deductions and net pay"""
    income_tax = 0.15 * gross_pay
    health_levy = 0.025 * gross_pay
    district_tax = 0.01 * gross_pay
    education_deduction = 0 if children <= 3 else (children - 3) * 0.5

    total_deductions = income_tax + health_levy + district_tax + education_deduction
    net_pay = gross_pay - total_deductions

    return income_tax, health_levy, district_tax, education_deduction, total_deductions, net_pay


print("=== WELCOME TO WAGEPAY PAYROLL SYSTEM ===")
print("Enter employee details. Type 'done' when finished.\n")

while True:
    name = input("Enter employee name (or 'done' to finish): ").strip()
    if name.lower() == "done":
        print("\nâœ… Payroll processing completed. Goodbye!\n")
        break

    # Get inputs
    try:
        hours = float(input(f"Enter hours worked for {name}: "))
        children = int(input(f"Enter number of children for {name}: "))
    except ValueError:
        print("âš ï¸ Invalid input. Please enter numeric values for hours and children.\n")
        continue

    # Calculations
    gross_pay = calculate_gross_pay(hours)
    income_tax, health_levy, district_tax, education_deduction, total_deductions, net_pay = calculate_deductions(gross_pay, children)

    # Formatted output
    print("\n=========================================")
    print(f"           PAYSLIP FOR {name.upper()}")
    print("=========================================")
    print(f"Hours Worked:                {hours}")
    print(f"Number of Children:          {children}")
    print(f"Gross Pay:                   â‚¦{gross_pay:.2f}\n")

    print("DEDUCTIONS:")
    print(f"  Income Tax (15%):          â‚¦{income_tax:.2f}")
    print(f"  Health Levy (2.5%):        â‚¦{health_levy:.2f}")
    print(f"  District Tax (1%):         â‚¦{district_tax:.2f}")
    print(f"  Education Deduction:       â‚¦{education_deduction:.2f}")
    print("-----------------------------------------")
    print(f"Total Deductions:            â‚¦{total_deductions:.2f}")
    print(f"Net Pay:                     â‚¦{net_pay:.2f}")
    print("=========================================\n")
````
